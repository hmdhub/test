```python
# import matplotlib
# print(matplotlib.__version__)
from matplotlib import pyplot as plt
import numpy as np
```

```python
# 日期从1991到2020,30年的日期
dates = np.arange(1991,2021)
# print(dates)
dates  # 输出
```

```python
# 使用numpy随机生成销量
sales = np.random.randint(50,500,size=len(dates))
sales
```

```python
# 绘制销量图
# x轴是数值类型，使用的是元素本身（int）
plt.xticks([dates[i] for i in range(0,len(dates),2)],rotation=45)
plt.plot(dates,sales)
```

```python
# 字符串类型
# x 轴
dates = np.arange(1991,2021).astype(np.str_) # astype(np.str_)指定数据类型（str_字符串）
# y 轴：销量
sales = np.random.randint(50,500,size=len(dates))
# 绘图
a = [i for i in range(0,len(dates),2)]
print(a)
plt.xticks([i for i in range(0,len(dates),2)],rotation=45)
plt.plot(dates,sales)
```

```python
"""
总结：
    1.x轴是数值型，会按照数值型本身作为x轴的坐标
    2.x轴为字符串类型，会按照索引作为x轴的坐标
"""
```

```python
"""
1.显示网络：plt.grid()

  显示网格
  linestyle：线型
  color：颜色
  linewidth：宽度
  axis：x，y，both，显示x/y两者之间的格网
"""
# x 轴 np.pi 圆周率
x = np.linspace(-np.pi,np.pi,endpoint=True)
# y 轴 
c = np.cos(x)
s = np.sin(x)
plt.plot(x,c,label='cos')
plt.plot(x,s,label='sin')
plt.legend()
plt.grid(True,linestyle='-')
```

```python
"""
2.plt.gca()对坐标轴的操作
"""
# 一.获取原图
#x轴
x = np.arange(-50,51)
# y 轴
y = x ** 2
# 绘图
plt.plot(x,y)
```

```python
# 二.目标：去掉上面和右边的边框
#x轴
x = np.arange(-50,51)
# y 轴
y = x ** 2
# 获取当前的边框
a = plt.gca()
# 通过spines，确定top，bottom，left，right（上，下，左，右）
a.spines['top'].set_color('none')# 去掉上面的边框
a.spines['right'].set_color('none')  # 去掉右面的边框
# 绘图
plt.plot(x,y)
```

```python
# 三.目标：移动左面和下面的边框（x，y轴）
# x轴
x = np.arange(-50,51)
# y 轴
y = x ** 2
# 获取当前的边框
a = plt.gca()
# 通过spines，确定top，bottom，left，right（上，下，左，右）
a.spines['top'].set_color('none')# 去掉上面的边框
a.spines['right'].set_color('none')  # 去掉右面的边框

# position 位置参数有三种：data，outward（向外移动），axes
a.spines['left'].set_position(('axes',0.5))  # axes参数范围[0.0 - 1.0]: 表示整个轴上的比例
a.spines['bottom'].set_position(('data',500))   # data参数表示按数值移动，给的数值代表挪动到y轴的刻度值
# 绘图
plt.plot(x,y)
```

```python
"""
plt.rcParams设置画图的分辨率，大小等信息

"""
# 绘制的原图
x = np.arange(-50,51)
y = x ** 2
plt.plot(x,y)
```

```python

x = np.arange(-50,51)
y = x ** 2
# plt.plot(x,y)
# # 设置分辨率
# plt.rcParams['figure.figsize']=(6.0,4.0) # 指定dpi=100，可以设置图片的大小
plt.rcParams['figure.figsize']=(7,10)
plt.plot(x,y)
```

```python
# 线条样式
# color：可以使用颜色的16进制，有可以使用线条颜色的英文，还可以使用之前的缩写
# alpha：0-1 透明度
# linestyle：折现样式
# marker标记点：o 圆圈标记；x x标记；D 钻石标记；H 六角标记；s 正方形标记；+ 加号标记
x = np.arange(0,100,10)
y = x ** 2
# color=''颜色，marker='H’标记点，linewidth='2
plt.plot(x,y,linewidth='2',label='test',color='b',linestyle='-',marker='H')
plt.legend(loc='upper left')
```

```python
# 线条样式缩写
# 颜色 标记 样式
plt.plot([1,2,3],[4,7,6],'r*-.')
# r 颜色；* 标记；—样式
```

```python
x = np.linspace(0,10,100)
# x
# 画图
plt.plot(x,x+0,'-g',label='-g') # 实线 绿色
plt.plot(x,x+1,'--c',label='--c') # 虚线 浅蓝色
plt.plot(x,x+2,'-.k',label='-.k')  #  点划线 黑色
plt.plot(x,x+3,'o',label='o') # 点  默认蓝色
plt.plot(x,x+4,'dr',label='dr') # 砖石 红色
```
