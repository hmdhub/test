```python
from matplotlib import pyplot as plt
import numpy as np
import matplotlib
print(matplotlib.__version__)  # 显示版本
```

```python
# 1.中文配置
plt.rcParams['font.sans-serif']=['SimHei']
# 负号显示
plt.rcParams['axes.unicode_minus']=False
```

```python
"""
柱状图的绘制:
   matplotlib.pyplot.bar(x,height,width:float = 0.8,bottom = None, *,
   align: str = 'center',data = None,**kwargs)
"""

# x 轴数据
x = range(5)
# y 轴数据
data = [5,20,15,25,30]
# 1.中文配置
plt.rcParams['font.sans-serif']=['SimHei']
# 负号显示
plt.rcParams['axes.unicode_minus']=False
# 设置标题
plt.title('基本柱状图')
# 网络
plt.grid(ls='--',alpha=0.5)
# 绘制柱状图
# plt.bar(x,data)
# bottom 参数
plt.bar(x,data,bottom = [10,5,20,25,30])  # 柱状图的起始点（橙色部分）
```

```python
# x 轴数据
x = range(5)
# y 轴数据
data = [5,20,15,25,30]
# 1.中文配置
plt.rcParams['font.sans-serif']=['SimHei']
# 负号显示
plt.rcParams['axes.unicode_minus']=False
# 设置标题
plt.title('设置颜色')
# 网络
plt.grid(ls='--',alpha=0.5)
# 绘制柱状图 设置颜色
plt.bar(x,data,color='pink')
```

```python
# facecolor 和 color 区别：facecolor 一种颜色；color多种颜色
# x 轴数据
x = range(5)
# y 轴数据
data = [5,20,15,25,30]
# 1.中文配置
plt.rcParams['font.sans-serif']=['SimHei']
# 负号显示
plt.rcParams['axes.unicode_minus']=False
# 设置标题
plt.title('设置颜色')
# 网络
plt.grid(ls='--',alpha=0.5)
# 绘制柱状图 设置颜色
plt.bar(x,data,color=['r','g','b'])
```

```python
"""
描边 - 想换的关键字参数为：
     edgecolor 或 ec（边缘线条的颜色）
     linestyle 或 ls
     linewidth 或 lw
"""
data = [5,15,20,25,30]
plt.title('设置边缘线条样式')
plt.bar(range(len(data)),data,ec='g',ls='--',lw=1,facecolor='pink')
```

```python
"""
同位置多个柱状图
"""
# 国家（x轴名称）
countries = ['挪威','德国','中国','美国','瑞典']
# 金牌
gold_medal = [16,9,12,15,9]
# 银牌
silver_mesal = [11,10,8,1,5]
# 铜牌
bronze_medal = [10,9,5,10,4]
# 设置宽度
width = 0.3
# 绘图
plt.bar(countries,gold_medal,color = 'gold',width=width)
plt.bar(countries,silver_mesal,color = 'green',width=width)
plt.bar(countries,bronze_medal,color = 'blue',width=width)

```

```python

# 国家（x轴名称）
countries = ['挪威','德国','中国','美国','瑞典']
# 金牌
gold_medal = [16,9,12,15,9]
# 银牌
silver_medal = [11,10,8,1,5]
# 铜牌
bronze_medal = [10,9,5,10,4]
# 1.将x轴转换为数值
x = np.arange(len(countries))
print(x)
width = 0.3
# 确定x的起始位置
gold_x = x  # 金牌
silver_x = x + width  # 银牌
bronze_x = x + 2 * width  # 铜牌
# 绘制图形
plt.bar(gold_x,gold_medal,width=width,color='gold')
plt.bar(silver_x,silver_medal,width=width,color='silver')
plt.bar(bronze_x,bronze_medal,width=width,color='brown')
# 将 x 轴转回到国家名字
# countries = ['挪威','德国','中国','美国','瑞典']
plt.xticks(x+width,labels=countries)

 # 显示高度文本
for i in range(len(countries)):
    # 金牌文本  va：水平  ha：垂直
    # verticalalignment: （va）垂直对齐方式，参数：['center' \ 'top' \ 'bottom' \'baseline']
    # horizontalalignment: (ha)水平对齐方式，参数：['center' \ 'right' \ 'left']
    plt.text(gold_x[i],gold_medal[i],gold_medal[i],va='bottom',ha='center')
    plt.text(silver_x[i],silver_medal[i],silver_medal[i],va='bottom',ha='center')
    plt.text(bronze_x[i],bronze_medal[i],bronze_medal[i],va='bottom',ha='center')
    
    
```

```python
# 绘制各国的金牌榜 银牌榜 铜牌榜和总排榜
# 国家（x轴名称）
countries = ['挪威','德国','中国','美国','瑞典']
# 奖牌总数
all_medal = [37,28,25,26,18]
# 金牌
gold_medal = [16,9,12,15,9]
# 银牌
silver_medal = [11,10,8,1,5]
# 铜牌
bronze_medal = [10,9,5,10,4]
# 创建绘图对象
fig = plt.figure(figsize=(6,4),dpi=70)
# 创建一个 2*2
ax1 = fig.add_subplot(221) # 位置为221
ax1.set_title('总牌榜')
# 旋转 x 轴标签
ax1.tick_params(axis='x',rotation=45)
ax1.bar(countries,all_medal,color='r')

# 绘制金牌榜图像
ax2 = fig.add_subplot(222)
ax2.set_title('金牌榜')
# 旋转 x 轴标签
ax2.tick_params(axis='x',rotation=45)
ax2.bar(countries,gold_medal,color='yellow')

# 绘制银牌榜图形
ax3 = fig.add_subplot(223)
ax3.set_title('银牌榜')
# 旋转 x 轴标签
ax3.tick_params(axis='x',rotation=45)
ax3.bar(countries,silver_medal,color='green')

# 绘制铜牌榜图形
ax4 = fig.add_subplot(224)
ax4.set_title('铜牌榜')
# 旋转 x 轴标签
ax4.tick_params(axis='x',rotation=45)
ax4.bar(countries,bronze_medal,color='pink')

```

```python
# 堆叠柱状图

# 国家（x轴名称）
countries = ['挪威','德国','中国','美国','瑞典']
# 奖牌总数
# all_medal = [37,28,25,26,18]
# 金牌
gold_medal = np.array([16,9,12,15,9])
# 银牌
silver_medal = np.array([11,10,8,1,5])
# 铜牌
bronze_medal = np.array([10,9,5,10,4])

# 绘制堆叠图
width = 0.3

plt.bar(countries,gold_medal,color='gold',label='金牌',bottom=silver_medal + bronze_medal,width=width)
plt.bar(countries,silver_medal,color='silver',label='银牌',bottom= bronze_medal,width=width)
plt.bar(countries,bronze_medal,color='brown',label='铜牌',width=width)
plt.legend()  # 图例
plt.ylabel('奖牌数')  # y轴名字

# 设置文本值
for i in range(len(countries)):
    max_y = gold_medal[i]+silver_medal[i]+bronze_medal[i]
    plt.text(countries[i],max_y,max_y,va='bottom',ha='center')
```

```python
# 绘制柱状图
# 总电影
movies = ['千里千寻','玩具总动员4','黑衣人：全球追缉']
x = np.arange(len(movies))
# 第一天
real_day1 = [4053,7548,6543]
# 第二天
real_day2 = [1840,4013,3421]
# 第三天
real_day3 = [2080,1673,2342]

# 绘制柱状图
width = 0.3

# 确定x的起始位置
real_day1_x = x  
real_day2_x = x + width  
real_day3_x = x + 2 * width

# 绘制柱状图
plt.bar(real_day1_x,real_day1,width=width,color='pink')
plt.bar(real_day2_x,real_day2,width=width,color='m')
plt.bar(real_day3_x,real_day3,width=width,color='c')
plt.xticks(x+width,labels=movies)

for i in range(len(movies)):
    plt.text(real_day1_x[i],real_day1[i],real_day1[i],va='bottom',ha='center')
    plt.text(real_day2_x[i],real_day2[i],real_day2[i],va='bottom',ha='center')
    plt.text(real_day3_x[i],real_day3[i],real_day3[i],va='bottom',ha='center')
    
```
