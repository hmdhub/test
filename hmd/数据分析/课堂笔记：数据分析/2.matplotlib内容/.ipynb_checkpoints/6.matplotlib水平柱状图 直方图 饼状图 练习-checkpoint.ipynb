```python
# 绘制水平堆叠图
movies = ['新蝙蝠侠','阻击手','奇迹笨小孩']
# 转化ndarray
real_day1 = np.array([4053,2548,1543])
real_day2 = np.array([4053,2548,1543])
real_day3 = np.array([8080,3673,1342])

# 确定距离左侧
left_day2 = real_day1  # 第二天的距离左侧为第一天的数值
left_day3 = real_day2 + real_day1
                   
# 设置线条高度
height = 0.3

# 绘制图形
plt.barh(movies,real_day1,height=height)
plt.barh(movies,real_day1,left=real_day1,height=height)
plt.barh(movies,real_day1,left=real_day2 + real_day1,height=height)


# 设置文本数据
sum_data = real_day1 + real_day2 + real_day3
for i in range(len(movies)):
# ha：控制文本x位置参数表示文本边界框的左边、右边或中间
# va：控制文本的y位置参数，表示文本边界框的底部、中心或顶部
    plt.text(sum_data[i],movies[i],sum_data[i],va='center',ha='left')
# plt.xlim(0,sum_data.max()+2000)
                     
                     

```

```python

# 国家（x轴名称）
countries = ['挪威','德国','中国','美国','瑞典']
# 金牌
gold_medal = np.array([16,9,12,15,9])
# 银牌
silver_medal = np.array([11,10,8,1,5])
# 铜牌
bronze_medal = np.array([10,9,5,10,4])

# 给个宽度
width = 0.3

# 确定距离左侧
left_silver = gold_medal  # 第二天的距离左侧为第一天的数值
left_bronze = silver_medal + gold_medal

# 绘制图形
plt.barh(countries,gold_medal,height=height,color='gold',label='金牌')
plt.barh(countries,silver_medal,left =left_silver,color='silver',height=height,label='银牌')
plt.barh(countries,bronze_medal,left = left_bronze,color = 'pink',height=height,label='铜牌')
plt.legend()

# 设置文本数据
sum_data = gold_medal +silver_medal + bronze_medal

 # 显示高度文本
for i in range(len(countries)):
    plt.text(sum_data[i],countries[i],sum_data[i],va='center',ha='left')
  
```

```python
# 同一位置多个柱状图
movies = ['新蝙蝠侠','阻击手','奇迹笨小孩']
# 转化ndarray
real_day1 = np.array([4053,2548,1543])
real_day2 = np.array([4053,2548,1543])
real_day3 = np.array([8080,3673,1342])
# y轴转数值
num_y = np.arange(len(movies))
# 设置高度
height=0.3
# 计算每个图形高度的起始位置
movie1_start_y = num_y  # 第一个
movie2_start_y = num_y + height  # 第二个
movie3_start_y = num_y + 2 * height  # 第三个

# 绘图
plt.barh(movie1_start_y,real_day1, height=height,label='蝙蝠侠')
plt.barh(movie2_start_y,real_day2, height=height,label='阻击手')
plt.barh(movie3_start_y,real_day3, height=height,label='奇迹笨小孩')

# 插入标签（图例）
plt.legend()

# 转化y轴
plt.yticks(num_y+height,movies)

# 显示文本信息
for i in range(len(movies)):
    plt.text(real_day1[i],movie1_start_y[i],real_day1[i],va='center',ha='left')
    plt.text(real_day2[i],movie2_start_y[i],real_day2[i],va='center',ha='left')
    plt.text(real_day3[i],movie3_start_y[i],real_day3[i],va='center',ha='left')
```
