```python
# 5.图例legend（）
times = ['2015/8/1','2015/10/12','2015/12/23','2016/3/4','2016/4/5','2016/5/15','2016/7/26','2016/8/29','2016/12/2']
# 支出
expense = np.random.randint(300,1500,size=len(times))
# 收入
income = np.random.randint(600,2000,size=len(times))
plt.title('收入-支出图')
plt.xlabel('日期')
plt.ylabel('金额')
# 定制x轴
plt.xticks(range(len(times)),rotation=45)
plt.plot(times,expense,label='支出')  # label='支出'：右上角显示小方框支出
plt.plot(times,income,label='收入') # 都是随机生成，每一次运行图都不一样
plt.legend(loc='upper right')
```

```python
# 图例
plt.rcParams['font.sans-serif']=['SimHei']  # 设置中文
plt.rcParams['axes.unicode_minus']=False  # 使负数展现出来
# 日期
time = ['2015/8/7','2015/10/12','2015/12/23','2016/3/4','2016/5/15','2016/7/26','2016/10/6','2016/12/17']
# 支出
espense = np.random.randint(300,1500,size=len(time))
# 收入
incom = np.random.randint(500,2000,size=len(time))
plt.title('收入-支出图形',fontsize=40) # fontsize=40：设置字体大小 ---给表格起名字
plt.xlabel('年份',fontsize=20)  # 给x轴起名
plt.ylabel('金额',fontsize=20)  # 给y轴起名
# 定制x轴
plt.xticks(range(len(times)),rotation=45,color='blue')  # rotation=45 倾斜45度,color='blue'设置颜色
plt.plot(times,expense,label='支出')     # 日期和支出的关系   --label=“支出”：图例会直接识别并调用
plt.plot(times,income,label='收入') # 日期和收入的关系  --label=“收入”：图例会直接识别并调用
plt.legend()
plt.show()
```

```python
"""
一.绘制曲线图，要求如下：
        1.显示图形名称
        2.为x轴和y轴定义说明信息
        3.使用中文显示
        
二.已知数据为2000年到2020年20年之间的销售情况
   sales = [109,150,172,260,273,333,347,393,402,446,466,481,499,504,513,563,815,900,930,961]要求：
   1.绘制年度销量线形图x轴为年份y轴为销量
   2.显示图例
"""
plt.rcParams['font.sans-serif']=['SimHei']  # 设置中文
plt.rcParams['axes.unicode_minus']=False  # 使负数展现出来
# 年份
times = np.arange(2000,2020).astype(np.str_) # 设置数据类型astype(np.str_)
# 销量
sales = [109,150,172,260,273,333,347,393,402,446,466,481,499,504,513,563,815,900,930,961]
plt.title('年度销量线形图',fontsize=20)
# 绘图
# 设置x轴
plt.xlabel('日期',fontsize= 10) # x轴
# 设置y轴
plt.ylabel('销量',fontsize=10)  # y轴
plt.xticks(range(len(times)),rotation=45)
plt.plot(times,sales,label='年销量') 
plt.legend(loc='upper right')  # 图例legend
for x,y in zip(times,sales):
    plt.text(x,y,'%s个'%y)


```
