```python
from matplotlib import pyplot as plt
import numpy as np
# 设置中文
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False
# 设置分辨率
plt.rcParams['figure.dpi']=100
plt.rcParams['figure.figsize']=(5,3)
```

```python
# 年份
times = np.arange(2000,2020).astype(np.str_)
# 销量
sales = [109, 150, 172, 260, 273, 333, 347, 393, 402, 446, 466, 481, 499,504, 513, 563, 815, 900, 930, 961]
plt.title('年度销量线形图')
plt.xlabel('年份')
plt.ylabel('销量')
plt.plot(times,sales,label='销量')
# 增加图例
plt.legend()
```

```python
# 国家
countries = ['挪威','德国','中国','美国','瑞典']
# 金牌
gold_medal = [16,12,9,8,8]
# 银牌
silver_medal = [8,10,4,10,5]
# 铜牌
bronze_medal=[13,5,2,7,5]

# 设置宽度
width = 0.3
# 绘图
plt.bar(countries,gold_medal,color='gold',label='金牌',width=width)
plt.bar(countries,silver_medal,color='silver',label='银牌',width=width)
plt.bar(countries,bronze_medal,color='brown',label='铜牌',width=width)
plt.ylabel('奖牌数')
plt.legend()


```
