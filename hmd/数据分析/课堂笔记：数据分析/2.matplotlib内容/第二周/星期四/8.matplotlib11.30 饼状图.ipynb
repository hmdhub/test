```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
# 1.中文配置
plt.rcParams['font.sans-serif']=['SimHei']
# 负号显示
plt.rcParams['axes.unicode_minus']=False
```

```python
"""
饼状图pie（）
pyplot.pie(x,explode=None,labels=None,colors=None,autopct=None)
x:数组序列，数组元素对应扇形区域的数量大小
labels：列表字符串序列，为每个扇形区域备注一个标签名字
colors：设置颜色，默认按照颜色周期自动设置
autopct：格式化字符串"fnt%pct",使用百分比的格式设置每个扇形区的标签，并将其放置在扇形区内
pctdistance：设置百分比标签与圆心的距离
labeldistance：设置各扇形标签（图例）与圆心的距离
explode：指定饼图某些部分的突出显示，既呈现爆炸式
shadow：是否添加饼图的阴影效果
"""
```

```python
# 定义标签
labels = ['娱乐','育儿','饮食','房贷','交通','其他']
# 设置每个标签所占数量
x = [200,500,1200,7000,200,900]
# 绘制饼状图
plt.pie(x,labels=labels)
```

```python
"""
显示百分比percentage
autopct='%.2f%%
"""# 定义标签
labels = ['娱乐','育儿','饮食','房贷','交通','其他']
# 设置每个标签所占数量
x = [200,500,1200,7000,200,900]
# 绘制饼状图
plt.pie(x,labels=labels,autopct='%.2f%%')
```

```python

"""
饼状图的分离：
            explode：指定饼图某些部分的突出显示
"""
# 定义标签
labels = ['娱乐','育儿','饮食','房贷','交通','其他']
# 设置每个标签所占数量
x = [200,500,1200,7000,200,900]
# 标题
plt.title('家庭支出')
# 设置饼状图分离
explode = (0.03,0.05,0.06,0.04,0.08,0.021)
# 绘制饼状图  %.2f%%：表示显示百分比，保留两位小数
plt.pie(x,labels=labels,autopct='%.2f%%',explode=explode) # explode=explode 图之间有间距
```

```python

"""
设置饼状图百分比和文本距离中心位置：
         pctdistance：设置百分比标签与圆心的距离；
         labeldistance：设置各扇形标签（图例）与圆心的距离
"""
# 定义标签
labels = ['娱乐','育儿','饮食','房贷','交通','其他']
# 设置每个标签所占数量
x = [200,500,1200,7000,200,900]
# 标题
plt.title('家庭支出')
# 设置饼状图分离
explode = (0.03,0.05,0.06,0.04,0.08,0.04)
# 绘制饼状图   %.2f%%：表示显示百分比，保留两位小数
plt.pie(x,labels=labels,autopct='%.2f%%',
       explode=explode,labeldistance=1.35,
       pctdistance =1.2)
```
