```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
# 1.中文配置
plt.rcParams['font.sans-serif']=['SimHei']
# 负号显示
plt.rcParams['axes.unicode_minus']=False
```

```python
x，y：散点的坐标
s：散点的面积
c ：散点的颜色
marker:散点样式（默认值为实心圆，‘o’，其余样式同plt.plot()）
alpha：散点透明度（[0,1]之间的数，0表示完全透明，1表示完全不透明）
linewidths：散点的边缘线宽
edgecolors：散点的边缘颜色
cmap：Colormap：默认None，标量或者是一个colormap的名字，只有c是一个浮点数数组时才使用
```

```python
# 以下实例scatter（）函数接受长度相同的数组参数，一个用于x轴的值，另一个是y轴的值
# x 
x = np.array([1,2,3,4,5,6,7,8])
# y
y = np.array([1,4,9,16,7,11,23,18])
# 绘制散点图
plt.scatter(x,y)
```

```python
"""
设置图标大小
"""
# x 
x = np.array([1,2,3,4,5,6,7,8])
# y
y = np.array([1,4,9,16,7,11,23,18])
# 绘制散点图
s = (20 * np.random.randn(8) ** 2)
plt.scatter(x,y,s,alpha=0.7)
```

```python
"""
自定义点的颜色和透明度

"""
x = np.random.rand(50)
y = np.random.rand(50)
# 设置大小
s = (10 * np.random.randn(50))** 2
colors = np.random.rand(50)
plt.scatter(x,y,c=colors)
```
