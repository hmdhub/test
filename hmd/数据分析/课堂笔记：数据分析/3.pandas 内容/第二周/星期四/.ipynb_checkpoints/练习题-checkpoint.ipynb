```python
"""
2.如图创建DataFrames（4*4，值为0-100的随机数），通过索引得到以下值
  1）索引得到b，c列的所有值
  2）索引得到第三第四行的数据
  3）按顺序索引得到two，one行的值
  4）索引得到大于50的值
"""
a = np.random.randint(0,100)
print(a)
b = np.random.randint(0,100)
e = np.random.randint(0,100)
d = np.random.randint(0,100)
print(b,e,d)
data = {
    'a':[1,2,3,4],
    'b':[44.98,67.3,75.9,45],
    'c':[43.4,40.3,52,63.55],
    'd':[4,5,6,7]
}
info = pd.DataFrame(data,index = ['one','two','three','four'])
print('所有',info)
print('-------------------')
# 1）索引得到b，c列的所有值
print(info[['b','c']])
print('1-------------------')
# 2)第三行和第四行的数据
# info.loc[['three','four']]
print(info.iloc[2:])
print('2--------------------------------')
# 3)按照顺序索ne','two']]# 实现方法1
print(info.loc[['two','one']])
print('3****************************')


# 4)索引大于50的值
for i in info.values:
    for v in i:
        if v>50:
            print('索引',v)
print('4---------------------')  
# info.values > 50 还是info>50得到的是布尔值 true false，取值要用布尔值索引取值
num = info.values
print(num[num>50])
```

```python
"""
3.创建一个3*3，值在0-100区间随机值的DataFrames，分别按照index和第二列值大小，降序排序
"""
d = {
    'v1':pd.Series([1,2,3],index=['a','b','c']),
    'v2':pd.Series([2,3,4],index=['a','b','c']),
    'v3':pd.Series([3,4,5],index=['a','b','c']),
}
df = pd.DataFrame(d)
print(df)
print('--------------------')

# 按照index降序
df1 = df.sort_index(ascending=False)
print(df1)
print('--------------------')
# 按照第二列值大小排序
b = df.sort_values(by=['v2'],ascending=False)
print(b)
```
